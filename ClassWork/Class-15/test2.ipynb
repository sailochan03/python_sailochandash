```python
#Using tweep module
import tweepy
import keys

bearer_token = keys.BEARER_TOKEN
client = tweepy.Client(bearer_token=bearer_token)

user = client.get_user(username = 'sssmixxx',user_fields=['public_metrics'])
if user.data:
    print(f'Name of User: {user.data.name}')
    print(f'ID: {user.data.id}')
    print(f'Username: {user.data.username}')
    print(f'Followers Count: {user.data.public_metrics["followers_count"]}')
else:
    print('ERROR')
```

```python
query = 'Valorant'
tweets = client.search_recent_tweets(query=query, max_results=15)

for tweet in tweets.data:
    print(f"TWEET: {tweet.text}")
    print('\n###############################################################\n')
```
